```python
# 1. Setup: imports & find CSV in current folder (Colab)
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# 1a. show any csv files present
csvs = [f for f in os.listdir() if f.lower().endswith('.csv')]
print("CSV files found in current folder:", csvs)

# 1b. If nothing found, upload manually using the widget (run this cell to pick file)
if not csvs:
    from google.colab import files
    uploaded = files.upload()   # click "Choose Files" and upload your DisasterDeclarationsSummaries.csv
    csvs = list(uploaded.keys())

# 1c. Load the first csv found (change path if you want)
DATA_PATH = csvs[0]
print("Loading:", DATA_PATH)
df = pd.read_csv(DATA_PATH, low_memory=False)
print("Loaded shape:", df.shape)
df.head()
```

```python
# Basic structure
print("Columns:", len(df.columns))
display(df.info())
display(df.describe(include='all').T)

# Missing values summary
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)
display(pd.DataFrame({"missing_count": missing, "missing_pct": missing_pct}).head(40))
```

```python
# normalize column names to simple snake_case
df.columns = (df.columns
               .str.strip()
               .str.lower()
               .str.replace(r'\s+', '_', regex=True)
               .str.replace(r'[^0-9a-z_]', '', regex=True))
print("Cleaned columns:", df.columns.tolist()[:30])

# drop exact duplicates
before = len(df)
df = df.drop_duplicates()
print("Dropped duplicates:", before - len(df))
```

```python
# find columns containing 'date'
date_cols = [c for c in df.columns if 'date' in c]
print("Date-like columns found:", date_cols)

# parse them to datetime
for c in date_cols:
    df[c] = pd.to_datetime(df[c], errors='coerce')

# create a primary date feature if available (declaration date is common in FEMA data)
primary_date = None
for cand in ['declarationdate','incidentbegindate','incidentenddate','begin_date','end_date']:
    if cand in df.columns:
        primary_date = cand
        break
print("Primary date chosen:", primary_date)

if primary_date:
    df['year'] = df[primary_date].dt.year
    df['month'] = df[primary_date].dt.month
    df['weekday'] = df[primary_date].dt.day_name()
    display(df[[primary_date,'year','month']].head())
```

```python
# show top values for key categorical columns (adjust names as needed)
candidates = ['state','incidenttype','disastertype','typedeclared','title','incident']
for c in candidates:
    if c in df.columns:
        print(f"\n--- {c} (unique {df[c].nunique()}):")
        display(df[c].value_counts().head(10))
```

```python
import matplotlib.ticker as ticker

# 1) Top states
if 'state' in df.columns:
    plt.figure(figsize=(10,4))
    df['state'].value_counts().head(15).plot(kind='bar')
    plt.title("Top 15 states by number of declarations")
    plt.ylabel('Count'); plt.show()

# 2) Events per year
if 'year' in df.columns:
    yearly = df['year'].value_counts().sort_index()
    plt.figure(figsize=(10,4))
    yearly.plot(marker='o')
    plt.title("Events per Year")
    plt.ylabel('Count'); plt.xlabel('Year'); plt.show()

# 3) Incident types
if 'incidenttype' in df.columns:
    plt.figure(figsize=(10,4))
    df['incidenttype'].value_counts().head(20).plot(kind='bar')
    plt.title("Top incident types"); plt.show()
```

```python
# 1) Fill program-declared flags (0/1 columns) if present
flag_guess = [c for c in df.columns if 'programdeclared' in c or c.endswith('_program')]
print("Flag-like columns:", flag_guess)
for c in flag_guess:
    df[c] = df[c].fillna(0).astype(int)

# 2) Create duration_days if start and end dates exist
# heuristics: look for columns containing 'begin' and 'end' or incidentbegindate / incidentenddate
start_col = next((c for c in df.columns if 'begin' in c and 'date' in c), None)
end_col = next((c for c in df.columns if 'end' in c and 'date' in c), None)
if start_col and end_col:
    df['duration_days'] = (df[end_col] - df[start_col]).dt.days

# 3) Simplify long text columns: optionally truncate 'title' to first 80 chars for display
if 'title' in df.columns:
    df['title_short'] = df['title'].astype(str).str[:80]

# 4) Inspect cardinality of categorical cols
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
cat_card = {c: df[c].nunique() for c in cat_cols}
sorted(cat_card.items(), key=lambda x: -x[1])[:20]
```

```python
# numeric impute with median
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for c in num_cols:
    if df[c].isna().any():
        df[c] = df[c].fillna(df[c].median())

# categorical: fill with 'unknown'
for c in cat_cols:
    df[c] = df[c].fillna('unknown')
```

```python
# Choose small-cardinality categorical columns for one-hot
low_card = [c for c in cat_cols if df[c].nunique() <= 20]
print("Low-cardinality categorical (one-hot):", low_card)

df_enc = pd.get_dummies(df, columns=low_card, drop_first=True)

# For very high-cardinality categorical fields (like 'title') consider label encoding or dropping for modeling
print("Shape before:", df.shape, " -> after encoding:", df_enc.shape)
```

```python
from sklearn.feature_selection import VarianceThreshold

# Keep only numeric features for automatic filters
X = df_enc.select_dtypes(include=[np.number]).fillna(0)
print("Numeric feature count:", X.shape[1])

# 1) Remove zero-variance features
vt = VarianceThreshold(threshold=0.0)
vt.fit(X)
X_v = X.loc[:, vt.get_support()]
print("After variance filter:", X_v.shape[1])

# 2) Remove highly correlated features (|corr| > 0.9)
corr = X_v.corr().abs()
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
to_drop = [col for col in upper.columns if any(upper[col] > 0.90)]
print("To drop due to high correlation:", to_drop[:30], " (count:", len(to_drop), ")")
X_sel = X_v.drop(columns=to_drop)
print("Remaining features after correlation drop:", X_sel.shape[1])
```

```python
# find candidate target columns (adjust names if necessary)
candidate_targets = [c for c in df.columns if c in ('incidenttype','disastertype','typedeclared','incident')]
print("Candidate targets:", candidate_targets)

# If you have a target, run a quick RandomForest to get importances
if candidate_targets:
    target = candidate_targets[0]   # choose first candidate (change if you want)
    print("Using target:", target)
    y = df[target].astype('category').cat.codes  # label encode
    X_for_model = X_sel.copy()
    # remove target if accidentally encoded in features
    if target in X_for_model.columns:
        X_for_model = X_for_model.drop(columns=[target])
    from sklearn.model_selection import train_test_split
    from sklearn.ensemble import RandomForestClassifier
    X_train, X_test, y_train, y_test = train_test_split(X_for_model, y, test_size=0.2, random_state=42, stratify=y)
    rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
    rf.fit(X_train, y_train)
    importances = pd.Series(rf.feature_importances_, index=X_for_model.columns).sort_values(ascending=False)
    display(importances.head(30))
else:
    print("No candidate target found  skip supervised selection")
```

## Insights from Analysis

- *Top States*: Texas (4562 declarations), Missouri (2670), and Virginia (2460) had the highest number of disaster declarations.  
- *Yearly Trends*: The year 2020 had the maximum declarations (9388), followed by 2005 (4736), 2011 (2680), and 2008 (2300). This shows major peaks during these years due to severe disasters.  
- *Frequent Incident Types*: The most common disaster types were Severe Storms (16959), Hurricanes (11879), Floods (10389), Biological events (7610), and Snow-related disasters (3659).  
- *Missing Data*: Columns like enddate (and a few others) contained a high percentage of missing values, which were either filled or dropped depending on relevance.  
- *New Features*: Extracted year, month, and weekday from date columns, and created duration_days (difference between incident start and end dates). These features help analyze trends over time and severity of events.  
- *Selected Features for Modeling*: State, Year, Month, Incident Type, Program Declared flags, and Duration Days were chosen as the most useful predictors after handling missing values, encoding categories, and removing highly correlated features.
